# Measure of central tendency
* It means we are trying to find out single numeric value with which we can represent central point of our dataset or say central value over which complete data is distributed
    * **Mean:**
        * Average/Arithmetic mean
        * Mean of population represented as (µ)
        * Mean of sample represented as x-bar (x̄)
        * Can only be used with numeric data not with categorical data
        * Mean is sensitive to outliers
    * **Median:**
        * It represents exact middle point of a sorted dataset
        * It can be used if data is having outliers
        * Can only be used with numeric data not with categorical data
    * **Mode:**
        * It represents most frequent occuring value or category in dataset
        * Mostly used for categorical data
        * Data can have more than 1 mode
        * Can be used for both numerical and categorical data

## For Frequency vs Height or Salary graph
* **Normal distributed/Symmetric:** Mean = Median = Mode
* **Left Skewed:** Mean < Median < Mode (Mean is not able to represent central part of data). So, we will take help of median to represent central part of data.
* **Right Skewed:** Mean > Median > Mode (Mean is not able to represent central part of data). So, we will take help of median to represent central part of data.
* The formula for Skewness is:
    * **Sk = (Mean - Mode) / Standard Deviation**
* The sign of Sk gives the direction and its magnitude gives the extent of skewness.
    * For Sk > 0, the distribution is right-skewed.
    * For Sk < 0 the distribution is left-skewed.

# Measure of Spread/Variance
* It represents the amount of dispersion in a dataset i.e. how spread out are the values 
* How far away the data points tend to fall from the center.
* This type of measure only applies to ordinal and numeric data that can be ranked
* A measure of spread gives us an idea of how well the mean represents the data. 
* When a distribution has lower variability, the values in a dataset are more consistent. 
* When the variability is higher, the data points are more dissimilar and extreme values become more likely.

## There are 4 types of measure of spread
* **Range**
* **IQR (Interquartile Range)**
* **Mean Absolute Deviation**
* **Variance**
* **Standard Deviation**

## Range:
* Simplest measure of variation
* Difference between the largest and smallest value
* If the Range is high that means our data is not close to the mean value, therefore our data is inconsistent.
* If the Range is low that means our data is close to the mean value, therefore our data is maybe or may not be consistent.
* Does not represent actual data distribution
* Very sensitive to outliers
* Does not consider each value of data

## IQR
* IQR = Q3-Q1
* IQR is a measure of where the majority of the values lies.
* Quartiles tell us about the spread of a data set by breaking the data set into quarters, just like the median breaks it in half
* Quartiles are a useful measure of spread because they are much less affected by outliers or a skewed data set than the equivalent measures
* Quartiles are often reported along with the median as the best choice of measure of spread and central tendency, when dealing with skewed and/or data with outliers

### So to represent our data correctly
* **Normally Distributed Data & No Outliers** => Mean and Range
* **Skewed & Outliers** => Median and IQR 
* With help of **Mean and Median** we will find the central point, and with help of **Range and IQR** we will find out how our values are distributed

## Mean Absolute Deviation
* It represents the amount of variation that occurs around the mean value in data set.
* Taking average of sum of absolute difference between each value of data set and mean
* **MAD = (i=1Σn|xi - µ|)/n**

## Variance (σ^2)
* The variance is a numerical value used to indicate how widely individuals in a group vary. It measures dispersion around the mean.
* If the values in data are spread out, the variance will be a large number. 
* Conversely, if the values are spread closely around the mean, the variance will be a smaller number
* **Variance (population) = Σ [(xi - μ)^2] / N**
* **Variance (sample) = Σ(xi−x̄)^2/(N-1)** 
* Variance is low (Spread of data is low), Variance is high (Spread of data is high)
* Problems in using variance - 
    * Gives more weight to extreme scores
    * Variance is not in the same units as the values in our data set, variance is measured in the units squared
* If data is having outliers then better not use variance

## Standard Deviation (σ)
* It is the measurement of average distance between each quantity and mean. That is, how data is spread out from mean
* It is square root of variance.
* A low standard deviation value indicates that the data points tend to be close to the mean, while a high value indicates that the data points are spread out over a wider range of values
* It'll solve the above problem that we were facing using variance

# Measure of Position
* **Percentile:**
    * Let suppose you gave an exam and scored 80%, that means your percentage is 80% out of 100% this is calculated with respect to total marks. But let suppose you got 80% percentile then that means you have scored more than 80% of the total student
    * Median is 50th percentile
* **Quartile:**
    * It divides the sorted data in quarters just like median does, but it'll divide in multiple quarters.
    * **Q1:** 25th Percentile
    * **Q2:** 50th percentile/Median
    * **Q3:** 75th percentile
* **Standard Score (z-score):**
    * A standard score indicates how many standard deviations an element is away from the mean. 
    * It helps in detecting outliers
    * Z-score is a measure how much the data differs from mean. We can determine whether it is different enough to be significant
    * **z-score = (xi - x̄)/Standard_deviation**
    * To calculate a z-score, we take the individual value and subtract the mean and then divide this difference by the standard deviation.
    * If z is negative then x is below average
    * If z is 0 then x is equal to the average
    * If z is positive then x is above the average
