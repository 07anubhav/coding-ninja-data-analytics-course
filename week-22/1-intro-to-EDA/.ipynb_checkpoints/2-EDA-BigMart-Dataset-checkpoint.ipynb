```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Understanding Data

```python
data = pd.read_csv('bigmart.csv')
data.head()
```

```python
data.head(-2)
```

# Data type and Conversions 
* **df['col'].astype(datatype)**: is used to convert the data type of column

```python
data.info()
```

```python
df = pd.DataFrame({'col1': ['1', '2'], 'col2': ['a','b'],'col3':['True','False']})
df.head()
```

```python
df.info()
```

### Change Data Type of column

```python
df['col1'] = df['col1'].astype(int)
```

```python
df.info()
```

```python
df['col3'] = df['col3'].astype(bool)
```

```python
df.info()
```

```python
data['Outlet_Establishment_Year'] = data['Outlet_Establishment_Year'].astype(str)
data.info()
```

# Understanding Data from Graph
* **pairplot:** It'll take that particular coulmn and plot a graph with all numeric data type columns

```python
# plt.figure(figsize=(30,30))
sns.pairplot(data,y_vars='Item_Outlet_Sales',height=5)
```

```python
plt.figure(figsize=(30,10))
sns.barplot(data=data,x='Item_Type',y='Item_Outlet_Sales')
```

```python
plt.figure(figsize=(30,10))
sns.barplot(data=data,x='Outlet_Type',y='Item_Outlet_Sales')
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(data=data,x='Outlet_Type',y='Item_Outlet_Sales')
```
