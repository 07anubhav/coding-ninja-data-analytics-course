# Sports and Countries

```python
Write a program to read the dataset "olympics.csv" and create a dictionary that shows the total number of sports 
played by each country.
```

```python
data_dict = {}
with open('olympics.csv',mode='r') as obj:
    data = csv.reader(obj)
    next(data) 
        
    for row in data:   
        country = row[6].strip()   
        sport = row[13].strip()    
            
        if country in data_dict:
            data_dict[country] += 1
        else:
            data_dict[country] = 1

            
# Create a dictionary with total sport counts
total_sport_counts = {
    country: sports
    for country, sports in data_dict.items()
}

print(total_sport_counts)
```

# Athletes

```python
Write a program to read the file "olympics.csv" and print the total number of  unique names of athletes whose names start 
with the letter "A" and end with the letter "i".
```

```python
unique_names = set()
with open('olympics.csv',mode='r') as obj:
    data = csv.reader(obj)
    next(data)  # This will skip the header row
        
    for row in data:
        athlete_name = row[1]
            
        if athlete_name.startswith("A") and athlete_name.endswith("i"):
            unique_names.add(athlete_name)
            
print(len(unique_names))
```

# Sports played by each country

```python
Write a program that reads the "olympics.csv" file and prints the total number of sports each country plays. 
[only implement the given function]

Hint: Use the 'Team' and 'Event' columns
```

```python
def total_sports_per_country(csv_file):
    data_dict = {}
    with open(csv_file,mode='r') as obj:
        data = csv.reader(obj)
        next(data)  

        for row in data:   
            country = row[6].strip()  # Assuming the country names are in the 6th column (index count from 0)
            sport = row[13].strip()   # Assuming the sport names are in the 12th column (index count from 0)

            if country in data_dict:
                # This will add the value to key of the dict if key already exists and the value will be stored in form of set
                # It wont add duplicate records to value
                data_dict[country].add(sport) 
            else:
                # This is creating a set, where we will store the value of the sports played by a country
                data_dict[country] = {sport}

    total_sport_counts = {
        country: len(sports) 
        for country, sports in data_dict.items()
    }
    
    return total_sport_counts
    

csv_file = 'olympics.csv'

country_total_sports = total_sports_per_country(csv_file)
for country, total_sports in country_total_sports.items():
    print(f"{country}: {total_sports}")
```

# Unique City

```python
 Write a program that reads the "olympics.csv" file and prints the total number of unique city names.
```

```python
def get_unique_cities(csv_file):
    unique_names = set()
    with open(csv_file,mode='r') as obj:
        data = csv.reader(obj)
        next(data)  
        
        for row in data:
            city_name = row[11].strip()
            
            unique_names.add(city_name)
    
    return len(unique_names)

unique_cities = get_unique_cities('olympics.csv')
print(unique_cities)
```

# Country Name

```python
Given dataset "olympics.csv", write a program that prints the country that has participated in the maximum number of 
unique sports.

Hint: Use the Team and Sport columns
```

```python
csv_file = 'olympics.csv'
def find_country_with_max_sports(csv_file):
    data_dict = {}
    with open(csv_file,mode='r') as obj:
        data = csv.reader(obj)
        next(data)  

        for row in data:   
            country = row[6].strip()  
            sport = row[12].strip()    

            if country in data_dict:
                data_dict[country].add(sport) 
            else:
                data_dict[country] = {sport}

    total_sport_counts = {
        country: len(sports) 
        for country, sports in data_dict.items()
    }
    
    max_sports = 0
    for k,v in total_sport_counts.items():
        if v>max_sports:
            max_sports=v
            country=k
    
    return country,max_sports

most_sports_country, max_sports_count = find_country_with_max_sports('olympics.csv')
print(most_sports_country, max_sports_count)
```

# Unique sport

```python
You are given a CSV file named "olympics.csv" that contains information about various Olympic events. 
Each row represents an event and contains the following fields: "Year", "City", "Sport", "Discipline", and "Athlete". 
Write a Python program that reads the CSV file and performs the following task:

Print the names of the unique sports played across the globe in the Olympic events.
```

```python
def get_unique_sport(csv_file):
    unique_sport = set()
    with open(csv_file,mode='r') as obj:
        data = csv.reader(obj)
        next(data)  
        
        for row in data:
            sport_name = row[12].strip()
            
            unique_sport.add(sport_name)
    
    return len(unique_sport)

unique_sports = get_unique_sport('olympics.csv')
print(unique_sports)
```

# Gold Medalist

```python
Given a `olympics.csv` file write a program that reads the data and prints the country that has won the most gold medals of 
all time and returns the country's name along with the corresponding gold medal count. 
```

```python
def find_most_gold_medals_csv(csv_file):
    with open(csv_file,mode='r') as obj:
        data = csv.DictReader(obj) 
        file_list = list(data) 
    
    gold_medal = {}
    for i in file_list:
        key = i['Team']
        value = i['Medal']
        
        count = 1
        if value == 'Gold'.strip():
            if key in gold_medal:
                gold_medal[key] += count
            else:
                gold_medal[key] = count
    
    max_medal = 0
    team = None
    for k,v in gold_medal.items():
        if v>max_medal:
            max_medal=v
            team=k
    
    return team,max_medal

most_gold_country, max_gold_medals = find_most_gold_medals_csv('olympics.csv')
print(most_gold_country, max_gold_medals)
```
