# Splitting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bigmart.csv')
df['rowno'] = np.arange(len(df))
df.head()
```

### .filter() : it'll filter out speicific column from dataframe and create new dataframe

```python
df_items = df.filter(['rowno','Item_Identifier','Item_Weight','Item_Fat_Content','Item_Visibility','Item_MRP','Item_Outlet_Sales'])
df_items.head()
```

```python
df_outlets = df.filter(['rowno','Outlet_Identifier','Outlet_Establishment_Year','Outlet_Size','Outlet_Location_Type','Outlet_Type','Item_Outlet_Sales'])
df_outlets.head()
```

### .iloc[row,columns]

```python
df.iloc[:,[12,0,1,2,3,5,11]]
```

# Merging

```python
df = pd.read_csv('bigmart.csv')
df['rowno'] = np.arange(len(df))
df.head()
```

```python
pd.merge(df_items,df_outlets,on='rowno')
```

### how='inner'/'left'/'right' : it signifies we have used which kind of join to merge the dataframe

```python
pd.merge(df_items.iloc[:,[0,1,2,3,4,5,6]],df_outlets,on='rowno',how='inner')
```

```python
pd.merge(df_items.iloc[0:5,[0,1,2,3,4,5,6]],df_outlets,on='rowno',how='inner')
```

```python
pd.merge(df_items.iloc[0:5,[0,1,2,3,4,5,6]],df_outlets,on='rowno',how='left')
```

```python
pd.merge(df_items.iloc[0:5,[0,1,2,3,4,5,6]],df_outlets,on='rowno',how='right')
```

# Grouping

```python
df = pd.read_csv('bigmart.csv')
df['rowno'] = np.arange(len(df))
df.head()
```

```python
df.groupby('Outlet_Type').sum()
```

```python
df.groupby('Outlet_Type').agg({'Item_Outlet_Sales':np.sum})
```

```python
df.groupby(['Outlet_Type','Outlet_Location_Type']).agg({'Item_Outlet_Sales':np.sum})
```

```python
df.groupby(['Outlet_Type','Outlet_Size']).agg({'Item_Outlet_Sales':np.sum})
```

```python
df.groupby(['Outlet_Type','Outlet_Size']).agg({'Item_Outlet_Sales':np.mean})
```

```python
df.groupby(['Outlet_Type','Outlet_Size']).agg({'Item_Outlet_Sales':np.max})
```

### Analyzing statistics by group

following statistical functions are available at groupby dataframes level.

| Function | Description |
|:---------|:------------|
| mean() | Compute mean of groups |
| sum() | Compute sum of group values |
| size() | Compute group sizes |
| count() | Compute count of group |
| std() | Standard deviation of groups |
| var() | Compute variance of groups |
| sem() | Standard error of the mean of groups |
| describe() | Generates descriptive statistics |
| first() |  Compute first of group values |
| last() | Compute last of group values |
| nth() | Take nth value, or a subset if n is a list |
| min() | Compute min of group values |
| max() | Compute max of group values |
