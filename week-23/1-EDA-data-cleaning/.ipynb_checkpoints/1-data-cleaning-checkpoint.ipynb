# Missing Values
* Types of missing values:
    * **MCAR (Missing completely At Random):**
        * No relationship between the missing values to the corresponding values and features
        * Ex: Secondary email
    * **MAR (Missing At Random):**
        * Systematic relationship between the missing values and the other values and features
        * Ex: Weight, Height, Age of females
    * **MNAR (Missing Not At Random):**
        * Known relationship (it is missing as there might be some reason why it's missing)
        * Ex: Sick leave in office, dropping out from college

### Emperical Rule:
   * **if <5% is missing:** ignore the missing values
   * **if >40% is missing:** ignore and report to the team
   * **if more than 5% and less than 40%:** take decision according to missing data
       * MCAR/MNAR: ignore the missing values or do the best and worst case analysis
       * MAR: do imputation

## Missing values imputation

```python
df['rowno'] = np.arange(len(df))
df.head()
```

```python
df_sub = df.filter(['Item_Weight','Item_MRP','Item_Outlet_Sales','rowno'])
```

```python
df_sub.head()
```

```python
import statsmodels as sm
from statsmodels.imputation import mice # Multiple Imputation Chain Equation
```

#### mice.MICEData(): is used to apply mice algorithm over the dataset  

```python
df_m = mice.MICEData(df_sub)
```

#### update_all(no_of_iterationy it specify the iteration, for many iteration we need to apply the model needs to be re-run so that the missing values is imputed to the minimal value that will be best fit for missing values

```python
df_m.update_all(50)
```

```python
df_m
```

```python
x = df_m.data
x.head()
```

```python
df['Item_Weight'].plot(kind='hist')
df_m.plot_imputed_hist('Item_Weight')
plt.show()
```

#### After finding out the missing values using mice algo
#### we will join our data into the original dataset

```python
df.loc[df['rowno'].isin(x['rowno']),'Item_Weight'] = x['Item_Weight']
```

```python
df.info() # we can see for Item_Weight there is no null values
```
