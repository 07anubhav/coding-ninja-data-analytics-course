```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

a = [3,4,5,6,7]
b = [2,4,6,8,9]

plt.scatter(x,y)
plt.plot(x,y)

plt.scatter(a,b)
plt.plot(a,b)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

a = [3,4,5,6,7]
b = [2,4,6,8,9]

# 'r' represent colour, '--' represent dotted, there are various way to use it
plt.plot(x,y,'r--')

plt.plot(a,b,'y^')
plt.show()
```

### Plot x^3

```python
x = np.array([3,4,5,6,7])
y = x**3

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(2,10,0.5)
y = x**3

plt.plot(x,y,'bo')
plt.show()
```

```python
# Matplotlib will treat it as y and automatically it'll take x as [0,1,2]
# it'll create x as range(n) which will create x array from 0 to n-1
a = [1,2,3] 

plt.plot(a)
plt.show()
```
