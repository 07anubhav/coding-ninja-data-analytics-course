# Plotting Graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

a = [3,4,5,6,7]
b = [2,4,6,8,9]

plt.scatter(x,y)
plt.plot(x,y)

plt.scatter(a,b)
plt.plot(a,b)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

a = [3,4,5,6,7]
b = [2,4,6,8,9]

# 'r' represent colour, '--' represent dotted, there are various way to use it
plt.plot(x,y,'r--')

plt.plot(a,b,'y^')
plt.show()
```

### Plot x^3

```python
x = np.array([3,4,5,6,7])
y = x**3

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(2,10,0.5)
y = x**3

plt.plot(x,y,'bo')
plt.show()
```

```python
# Matplotlib will treat it as y and automatically it'll take x as [0,1,2]
# it'll create x as range(n) which will create x array from 0 to n-1
a = [1,2,3] 

plt.plot(a)
plt.show()
```

# Customizing Graphs

### color and marker

```python
x = np.arange(0,10,0.5)
y = x**3

plt.plot(x,y,color='black',marker='+')
plt.show()
```

### linewidth

```python
x = np.arange(0,10,0.5)
y = x**3

plt.plot(x,y,color='black',linewidth='5')
plt.show()
```

### y-label, x-label, legend, title

```python
x = np.arange(0,10,0.5)
y1 = x**3
y2 = x**2

plt.plot(x,y1,color='black',linewidth='3',label='x^3')
plt.plot(x,y2,color='yellow',linewidth='3',label='x^2')

plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.title('Matplotlib Demo')
plt.show()
```

### axis and grid
* for plt.axis(**[x-start, x-end, y-start, y-end]**)

```python
x = np.arange(0,10,0.5)
y1 = x**3
y2 = x**2

plt.plot(x,y1,color='black',linewidth='3',label='x^3')
plt.plot(x,y2,color='yellow',linewidth='3',label='x^2')

plt.axis([0,10,0,100])
plt.ylabel('y')
plt.xlabel('x')

plt.grid()
plt.legend()

plt.title('Matplotlib Demo')
plt.show()
```

### text
* plt.text(x,y,'text you want to show')

```python
x = np.arange(0,10,0.5)
y1 = x**3
y2 = x**2

plt.plot(x,y1,color='black',linewidth='3',label='x^3')
plt.plot(x,y2,color='yellow',linewidth='3',label='x^2')

plt.axis([0,10,0,100])
plt.ylabel('y')
plt.xlabel('x')

plt.text(2,80,'Text',fontsize=20)
plt.grid()
plt.legend()

plt.title('Matplotlib Demo')
plt.show()
```

# Bubble Chart

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

plt.scatter(year,salary,s=population,c='r')

plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
```

### alpha
* It determines the opacity of the bubble, by default it's 1. Less the value less the opacity

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

plt.scatter(year,salary,s=population,c='r',alpha=0.5)

plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
```

### set color of bubble

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

c = np.arange(len(year))
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black')

plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

c = np.arange(len(year))
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black',marker='^')

plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

c = np.arange(len(year))
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black',marker='^')

plt.text(year[0]+0.1,salary[0]+0.1,population[0])
plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
```
