```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Pie Chart

```python
size = [3,4,6,2]
color = ['blue','purple','red','yellow']
label = ['A','B','C','D']

plt.title('Split among classes')
plt.pie(size,colors=color,labels=label)
plt.show()
```

### autopct
* We can pass a function as well into it as what value to show in different regions. Otherwise it'll pick by it's own

```python
size = [3,4,6,2]
color = ['blue','purple','red','yellow']
label = ['A','B','C','D']

plt.title('Split among classes')
plt.pie(size,colors=color,labels=label,autopct='%.2f%%') # it means give percentage in float values, %% show the percentage sign
plt.show()
```

### explode
* it'll explode the region which we specify, that particular region will be cutoff from the main graph
* the graph will be in counter clockwise by default

```python
size = [3,4,6,2]
color = ['blue','purple','red','yellow']
label = ['A','B','C','D']
explode = [0.1,0,0,0]

plt.title('Split among classes')
plt.pie(size,colors=color,labels=label,autopct='%.2f%%',explode=explode) 
plt.show()
```

### start angel
* it will tell us from which angle to start our graph

```python
size = [3,4,6,2]
color = ['blue','purple','red','yellow']
label = ['A','B','C','D']
explode = [0.1,0.2,0,0]

plt.title('Split among classes')
plt.pie(size,colors=color,labels=label,autopct='%.2f%%',explode=explode,counterclock=False,startangle=90) 
plt.show()
```

# Histogram
* By default histogram can create 10 bins (towers)
* size of each bin is **((max_num-min_num)/num_of_bins)**
* It is also known as frequency graph

```python
a = [1,2,3,4,1,1,3,2,4,2,4,2,3,1,4,1,1,2,4,3,3,1]

plt.hist(a)
plt.show()
```

### xticks
* If the numbers are clubed together then xticsk will help us like how on x-axis numbers are distributed

* 22-1/10 = 2.1, so 2.1 is the bin width

```python
a = [1, 2, 1, 3, 5, 4, 5, 5, 4, 3, 3, 10, 7, 8, 8, 6, 7, 20, 22, 11, 14, 15]

plt.hist(a,edgecolor='black')
plt.xticks([1,2,3,4,5,6])
plt.show()
```

### bins
* calculated by = (21-1)/21 = 1, so we selected 21 so that each bin size will be of width 1

```python
a = [1, 2, 1, 3, 5, 4, 5, 5, 4, 3, 3, 10, 7, 8, 8, 6, 7, 20, 22, 11, 14, 15]
xt = np.arange(0,23)

plt.hist(a,edgecolor='black',bins=21)
plt.xticks(xt)
plt.show()
```

```python
a = [1, 2, 1, 3, 5, 4, 5, 5, 4, 3, 3, 10, 7, 8, 8, 6, 7, 20, 22, 11, 14, 15]
xt = np.arange(0,23)

plt.hist(a,edgecolor='black',bins=[1,3,6,10,15,20]) # We can specify the range in bins
plt.axis([0,25,0,10])
plt.xticks(xt)
plt.show()
```

# Bar Graph

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

plt.bar(year,population)
plt.show()
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

plt.bar(year,population,width=0.5) # width of bar
plt.show()
```

```python
year = ["year - 2012","year - 2013","year - 2014","year - 2015","year - 2016","year - 2017"]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

plt.bar(year,population,width=0.6,edgecolor='yellow')
plt.xticks(rotation=45) # otherwise xticks will be overlapped 
plt.show()
```

# How to decide graph type
* **Line**: When we have some few categories and how they compare with eachother (for large amount of data)
* **Scatter**: When we want to see the co-relation between 2 variables
* **Bubble**: When we have 2 varibales but we have 3rd variable as well and we feel it's an important variable
* **Bar**: When we have some few categories and how they compare with eachother (for less amount of data)
* **Pie**: When we want to show percentages among different features
* **Histogram**: We use this to see the frequency of the variable
