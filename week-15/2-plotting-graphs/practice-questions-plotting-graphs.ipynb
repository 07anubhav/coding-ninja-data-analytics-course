# Line Plot Unplaced Students

```python
Write a program to plot a Line Graph of the number of unplaced students versus the year.
Print the number of years when the number of unplaced students was greater than the previous year.
The dataset is a CSV file called "placement_record.csv".

Note: No. of unplaced students = Total no. of students - No. of students placed in software and data
Input:
    placement_record.csv(CSV file)
Output:
    num(int): An integer containing the number of years
```

```python
df = pd.read_csv('placement_record.csv')

df['Unplaced_student'] = df['Total students']-(df['students_placed_software']+df['Students_placed_Data'])

x = np.array(df['Year'])
y =np.array(df['Unplaced_student'])

plt.plot(x,y)
plt.grid()
plt.xlabel('Year')
plt.ylabel('Unplaced Student')

plt.show()

count = 0
for i in range(len(x)-1):
    if y[i+1]>y[i]:
        count += 1
        
print(count)
```

# Bubble Chart

```python
Write a program to plot a bubble chart with x-axis values, y-axis values, and bubble sizes.
Input:
    
Three lists:
x: The list of values for the x-axis.
y: The list of y-axis values for the bubble chart.
z: The list of bubble sizes representing the z-values.
    
Output:
The program should display a bubble chart with the specified x-axis values, y-axis values, and bubble sizes.
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

c = np.arange(len(year))
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black',marker='^')

for i in range(len(year)):
    plt.text(year[i]+0.3,salary[i]+0.1,population[i])

plt.axis([2010,2020,10,30])
plt.grid()
plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
```

# Laptop Sales

```python
Given data is having distribution of sales of different laptops for the year 2018 in India 
(it has the company name and number of laptops sold).
Plot a pie graph to show the %age distribution for the year 2018 and also print the value of %age distribution of individual 
companies.

Note: Output should have company name and %age distribution (%age value should be up to 1 decimal place without the % sign), 
it should be in the same order as it is given.
    
Output Format:
company_name1 percentage1
company_name2 percentage2
```

```python
company = ['HP','Dell','Lenovo','Asus','Apple','Acer']
sold = [20000,43000,15000,17000,22000,13000]

plt.pie(sold,labels=company,autopct='%.1f%%',counterclock=False,startangle=90)
plt.axis('equal')
plt.show()

sold_prec = [
    round((i/(sum(sold)))*100,1)
    for i in sold
]

for i in range(len(company)):
    print(f'{company[i]} {sold_prec[i]}')
```

# Pie Chart Products Sold

```python
Given a dictionary containing product names as keys and a list containing the number of products sold per day as values, 
write a program to plot a Pie Chart. 

The labels of the Pie Chart are the product names, and the size of each slice is the total number of products sold.

Print the name of the product that was bought the most.

Input:
product_sales: A dictionary where the keys represent the product names, and the values are lists containing the number of 
products sold per day.
Output:
product(String): Name of product
```

```python
product_sales={"AC":[12,23,45,23,56,78,90,109], "Phone":[10,40,23,13,67,9,45,9], "Fridge":[24,80,100,101,240,78,109,86]}

product = []
total_sale = []

for k,v in product_sales.items():
    product.append(k)
    total_sale.append(sum(v))

plt.pie(total_sale,labels=product,autopct='%.2f%%',counterclock=False,startangle=90)
plt.axis('equal')
plt.show()

print(product[total_sale.index(max(total_sale))])
```

# Pie Chart Major Sector

```python
Write a program to plot a pie chart where the labels represent the major sectors, and the size represents the total number 
of students hired in each major sector for all the years. 
Print the name of the sector where the majority of the students were placed.
The dataset is a CSV file called "placement_record.csv".

Input:
    placement_record.csv(CSV file)
Output:
    sector(string): Name of sector
```

```python
df = pd.read_csv('placement_record.csv')
df['Total_hired'] = df['students_placed_software']+df['Students_placed_Data']

major_sec = {}
sector = np.array(df['Major_sector'])

for i in range(11):
    if sector[i] in major_sec:
         major_sec[sector[i]] += df['Total_hired'].iloc[i]
    else:
        major_sec[sector[i]] = df['Total_hired'].iloc[i]

sector_array = []
value_array = []
print(major_sec)
for k,v in major_sec.items():
    sector_array.append(k)
    value_array.append(v)

plt.pie(value_array,labels=sector_array,autopct='%.1f%%',counterclock=False,startangle=90)
plt.show()

print(sector_array[value_array.index(max(value_array))])
```

# Repeated Words

```python
Write a program to plot a bar chart of the number of occurrences of unique words in a given string.
Print the number of words that occur more than once.
Input:
str1(String): Sentence
Output:
num(int): Number of words
```

```python
str1="Python is a high-levely with the use of also Python is dynamic and garbage-collected, thats Python"

l = str1.split(' ')

str_dict = {}
for i in l:
    str_dict[i] = str_dict.get(i,0)+1
        
word = []
freq = []

for k,v in str_dict.items():
    word.append(k)
    freq.append(v)
    
plt.bar(word,freq)
plt.xticks(rotation=90)
plt.show()
print(2)
```
